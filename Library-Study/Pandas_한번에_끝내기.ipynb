#### Index 연산

```python
idx1 = pd.Index([1, 2, 4, 6, 8])
idx2 = pd.Index([2, 4, 5, 6, 7])
print(idx1.append(idx2))  # 리스트 합치기
print(idx1.difference(idx2))  # 차집합
print(idx1 - idx2)  # 차집합과는 다르다
print(idx1.intersection(idx2))  # 교집합
print(idx1 & idx2)  # 교집합
print(idx1.union(idx2))  # 합집합
print(idx1 | idx2)  # 합집합
print(idx1.delete(0))  # 인덱스로 삭제
print(idx1.drop(1))  # 값으로 삭제
print(idx1 ^ idx2)  # 공통을 뺀 나머지 (A ∩ B)^c
```

### 고성능 연산

```python
# eval, query
```

```python
nrows, ncols = 100000, 100
df1, df2, df3, df4 = (pd.DataFrame(np.random.rand(nrows, ncols)) for _ in range(4))
```

```python
%timeit df1 + df2 + df3 + df4
```

```python
%timeit pd.eval('df1 + df2 + df3 + df4')  # 고성능 연산
```

```python
%timeit df1 * -df2 / (-df3 * df4)
```

```python
%timeit pd.eval('df1 * -df2 / (-df3 * df4)')
```

```python
%timeit (df1 < df2) & (df2 <= df3) & (df3 != df4)
```

```python
%timeit pd.eval('(df1 < df2) & (df2 <= df3) & (df3 != df4)')  # 오잉?
```

```python
df = pd.DataFrame(np.random.rand(1000000, 5), columns=['A', 'B', 'C', 'D', 'E'])
df.head()
```

```python
%timeit df['A'] + df['B'] / df['C'] - df['D'] * df['E']
```

```python
%timeit pd.eval('df.A + df.B / df.C - df.D * df.E')
```

```python
%timeit df.eval('A + B / C - D * E')  # column명으로 쓸 수도 있다.
```

```python
df.eval('R = A + B / C - D * E', inplace=True)  # inplace=True로 계산된 결과값을 반영할 수 있다.
df.head()
```

```python
col_mean = df.mean(1)
df['A'] + col_mean
```

```python
df.eval('A + @col_mean')  # @를 사용해서 외부에 있는 변수를 가져와서 eval에서 사용할 수 있다.
```

```python
df[((df.A < 0.5) & (df.B > 0.5) & (df.C > 0.5))]
```

```python
pd.eval('df[((df.A < 0.5) & (df.B > 0.5) & (df.C > 0.5))]')
```

```python
df.query('(A < 0.5) and (B > 0.5) and (C > 0.5)')  # df에 query를 날려서 조건이 포함된 상태를 질의 형태로 DataFrame에 날릴 수 있다.
```

```python
col_mean = df['D'].mean()
df[(df.A < col_mean) & (df.B < col_mean)]
```

```python
df.query('A < @col_mean and B < @col_mean')
```

```python
def create_df(cols, idx):
    data = {c: [str(c.lower()) + str(i) for i in idx] for c in cols}
    return pd.DataFrame(data, idx)
```

- `count` : 전체 개수

### 피벗 테이블(Pivot Table)

- `values` : 집계하려는 column 이름 혹은 이름의 list, 기본적으로 모든 숫자 컬럼 집계
- `index` : 피벗 테이블의 row를 그룹으로 묶을 column 이름이나 그룹 key
- `columns` : 피벗 테이블의 column을 그룹으로 묶을 column 이름이나 그룹 key
- `aggfunc` : 집계 함수나 함수 리스트, 기본값으로 `mean`이 사용
- `dropna` : True인 경우 모든 항목이 NA인 column은 포함하지 않음
- `margin` : 부분합이나 총계를 담기 위한 row/column 추가 여부, 기본값은 False

```python
df.pivot_table(['c3', 'c4'],
               index=['c1'],
               columns=['c2'])
```

```python
df.pivot_table(['c3', 'c4'],
               index=['c1'],
               columns=['c2'],
               margins=True)  # 부분합 총계를 추가
```

```python
df.pivot_table(['c3', 'c4'],
               index=['c1'],
               columns=['c2'],
               margins=True,
               aggfunc=sum)  # margins가 있다면 aggfunc도 가능
```

```python
df.pivot_table(['c3', 'c4'],
               index=['c1'],
               columns=['c2'],
               margins=True,
               aggfunc=sum,
               fill_value=0)  # 크으
```

```python
pd.crosstab(df.c1, df.c2)  # 간단하게 c1, c2에 대해서 table을 만들어줘.
```

```python
pd.crosstab(df.c1, df.c2, values=df.c3, aggfunc=sum, margins=True)
```

#### Null 값 처리

- `isnull()` : 누락되거나 NA인 값을 boolean 값으로 반환
- `notnull()` : 누락되지 않고 Na가 아닌 값을 반환
- `dropna()` : 누락된 데이터가 있는 축 제외
- `fillna()` : 누락된 값을 대체하거나 `ffill`이나 `bfill`로 보간 메서드 적용

```python
s = pd.Series([1, 2, np.nan, 'String', None])
s
```

```python
s[s.notnull()]
```
